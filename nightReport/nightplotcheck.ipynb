```python
%matplotlib inline

import lsst.sims.maf.db as db
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.metricBundles as metricBundles
import lsst.sims.maf.plots as plots
import lsst.sims.maf.utils as utils
```

```python
opsdb = utils.connectOpsimDb('minion_1016_sqlite.db')
```

```python
 # Plot up each visit
bundleList = []
night = 200
metric = metrics.NightPointingMetric()
slicer = slicers.UniSlicer()
sql = sql = 'night=%i' % night
plotFuncs = [plots.NightPointingPlotter()]
bundle = metricBundles.MetricBundle(metric, slicer, sql, plotFuncs=plotFuncs)
bundleList.append(bundle)
```

```python
bd = metricBundles.makeBundlesDictFromList(bundleList)
```

```python
group = metricBundles.MetricBundleGroup(bd, opsdb)
```

```python
group.runAll()
```

```python
group.plotAll(closefigs=False)
```
