```python
%matplotlib inline
import numpy as np

import lsst.sims.maf.utils as mafUtils
import lsst.sims.maf.db as db
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.metricBundles as metricBundles
import lsst.sims.maf.plots as plots

import matplotlib.pylab as plt
```

```python
databases = []
databases.append(db.Database('feature_baseline_10yrs.db', defaultTable='observations'))
databases.append(db.Database('feature_rolling_half_10yrs.db', defaultTable='observations'))
databases.append(db.Database('feature_rolling_half_mask_10yrs.db', defaultTable='observations'))
databases.append(db.Database('feature_rolling_third_10yrs.db', defaultTable='observations'))
databases.append(db.Database('feature_rolling_twoThird_10yrs.db', defaultTable='observations'))
names = ['baseline', 'half', 'half mask', 'third', 'two thirds']
#names = ['baseline', 'half mask']
```

```python
for name, database in zip(names, databases):
#for name, database in zip([names[0]], [databases[0]]):
    bundleList = []
    for year in [1,2,3,4,5,6]:
    #for year in [1]:
        slicer = slicers.HealpixSlicer(lonCol='RA', latCol='dec')
        metric = metrics.CountMetric(col='RA')
        sql='night > %i and night < %i' % ((year-1)*365.25, year*365.25)
        plotDict={'title':'%s, year %i' % (name, year), 'percentileClip':95.}
        bundleList.append(metricBundles.MetricBundle(metric, slicer, sql, plotDict=plotDict,
                                                     plotFuncs=[plots.HealpixSkyMap()], runName=name))
    bd = metricBundles.makeBundlesDictFromList(bundleList)
    mb = metricBundles.MetricBundleGroup(bd, database, outDir='Roll_plots')
    mb.runAll()
    mb.plotAll(closefigs=False)
```
