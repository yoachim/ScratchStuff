```python
%matplotlib inline
import lsst.sims.skybrightness as sb
import numpy as np
import healpy as hp
import matplotlib.pylab as plt
```

```python
nside = 32
lat, ra = hp.pix2ang(nside, np.arange(hp.nside2npix(nside)))
dec = np.pi/2-lat
```

```python
sm = sb.SkyModel(observatory='LSST', mags=True, zodiacal=False)
mjd = 59583.153342
sm.setRaDecMjd(ra,dec,mjd)
mag = sm.returnMags()
```

```python
sm.airmass
```

```python
for fn in ['u', 'g', 'r', 'i', 'z', 'y']:
    plt.figure()
    plt.plot(sm.airmass, mag[fn], 'ko')
    plt.xlabel('Airmass')
    plt.ylabel('$\mu_%s$ (mag arcsec$^{-2}$)' % fn)
```

```python
for fn in ['u', 'g', 'r', 'i', 'z', 'y']:
    hp.mollview(mag[fn], title=fn)
```
