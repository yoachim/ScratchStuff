```python
# just a quick grab of the opsim field locations
opsdb = db.OpsimDatabase('/Users/yoachim/Scratch/Opsim_sqlites/enigma_1189_sqlite.db')
```

```python
field_info = opsdb.fetchFieldsFromFieldTable()
```

```python
class dummySlicer(object):
    def __init__(self, ra, dec):
        self.slicePoints={'ra': ra, 'dec': dec}
    
# make a dummy slicer
slicer = dummySlicer(field_info['fieldRA'], field_info['fieldDec'])

# make a random map
data = np.random.rand(field_info.size)
# needs to be a masked array
data = ma.masked_array(data)
```

```python
plotter = plots.BaseSkyMap()
plotter(data, slicer, {'raCen': 0, 'colorMin': 0, 'colorMax':1, 'plotMask': True, 
                       'xlabel': 'some units', 'title': 'some title'})
```
